# This Notebook is to check the quality of the data and perform cleaup process for next steps

```python
#import necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# read csv data
TDCJ_Data1 = pd.read_csv("https://raw.githubusercontent.com/abirakm/Texas-Department-of-Criminal-Justice-Record/main/Texas_Department_of_Criminal_Justice_Receives_FY_2019.csv", dtype={'Offense Code': object},parse_dates=['Receive Date','Sentence Date','Offense Date'])
TDCJ_Data2 = pd.read_csv("https://raw.githubusercontent.com/abirakm/Texas-Department-of-Criminal-Justice-Record/main/Texas_Department_of_Criminal_Justice_Receives_FY_2020.csv", dtype={'Offense Code': object},parse_dates=['Receive Date','Sentence Date','Offense Date'])
TDCJ_Data = TDCJ_Data1.append(TDCJ_Data2)
TDCJ_Data.columns = TDCJ_Data.columns.str.replace(' ','_')
TDCJ_Data.columns = TDCJ_Data.columns.str.replace('(','')
TDCJ_Data.columns = TDCJ_Data.columns.str.replace(')','')
TDCJ_Data = TDCJ_Data.rename(columns={"Sentence_Years":"Sentence_Type"})
TDCJ_Data.County = TDCJ_Data.County.str.replace("Lasalle","La Salle")
TDCJ_Data.head()

```

For this analysis we will use the Gender, Race, Age,County,Offense Code, TDCJ Offense,Sentence Date, Offense Date and Sentence (Years)

```python
TDCJ_Data.info()
```

There are some null values in the Sentence Date. Let's remove those rows

```python
#Data Clean up drop null values for Sentence date and where sentece date is reported earlier than offence date

TDCJ_Data = TDCJ_Data.dropna()
TDCJ_Data =  TDCJ_Data.query('Offense_Date< Sentence_Date ')
TDCJ_Data.info()
```

As it is clear that the description is not consistent with the offence code. For simplycity lets consider the first two digits of the offence code and count how many unique offence type are there

```python
# make a copy of data for offence code and offence description
Offence_Code = TDCJ_Data[["Offense_Description","Offense_Code"]].copy()
Offence_Code["ShortCode"] =  TDCJ_Data["Offense_Code"].str.slice(0,2)
Offence_Code.head()
```

```python
Offence_Code = Offence_Code[["ShortCode","Offense_Description"]].\
groupby(["ShortCode","Offense_Description"]).\
size().\
reset_index(name="obs").\
sort_values(by =["ShortCode","obs"],ascending = [True, False]).\
groupby(["ShortCode"]).\
first().\
reset_index().drop("obs",axis =1)

Offence_Code
```

There are 37 unique offence type

```python
TDCJ_Data["ShortCode"] =  TDCJ_Data["Offense_Code"].str.slice(0,2)
TDCJ_Data =  TDCJ_Data.drop(["Offense_Description","Offense_Code"], axis =1)
TDCJ_Data =  pd.merge(TDCJ_Data,Offence_Code)
TDCJ_Data.head()
```

```python
TDCJ_Data["Case_Duration_Days"] =  (TDCJ_Data["Sentence_Date"]- TDCJ_Data["Offense_Date"])/np.timedelta64(1,"D")
TDCJ_Data["Sentence_Year"] = TDCJ_Data["Sentence_Date"].dt.year
TDCJ_Data["Offense_Year"] = TDCJ_Data["Offense_Date"].dt.year
TDCJ_Data["Age_at_Offence"] =  TDCJ_Data["Age"]-(TDCJ_Data["Sentence_Year"] - TDCJ_Data["Offense_Year"])
TDCJ_Data.drop(["Receive_Type","Sentence_Date","Offense_Date","Receive_Date", "Age","ShortCode"], axis =1,inplace=True)
TDCJ_Data.head()
```

Now e can merge with the socio demographic information with the the TDCJ dataset

```python
## Download US State data with socuo demographic Information
US_County = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/minoritymajority.csv')
State = "Texas"
Texas_County = US_County.query("STNAME == @State")
Texas_County.CTYNAME = Texas_County.CTYNAME.str.replace( " County","")
Texas_County['MalePCT'] = 100*Texas_County['TOT_MALE']/ Texas_County['TOT_POP']
Texas_County = Texas_County[['FIPS','CTYNAME','TOT_POP','MalePCT','BlackPCT','HispanicPCT']]
```

```python
# Define Function to rplace % and convert to numeric
def Change_PCT(PCT_STR):
    return PCT_STR.str.replace("%","")

## Change Black
Texas_County[["BlackPCT","HispanicPCT"]] = Texas_County[["BlackPCT","HispanicPCT"]].apply(Change_PCT).apply(pd.to_numeric, errors='coerce', axis=1)

Texas_County.head()
```

```python
Texas_County
```

```python
#Fix error in County Name
Texas_County.CTYNAME = Texas_County.CTYNAME.str.replace("DeWitt","Dewitt")
```

```python
#join TDCJ and County Demographic data

TDCJ_Data = TDCJ_Data.merge(Texas_County, how =  'left', left_on = 'County', right_on= 'CTYNAME' ).drop('CTYNAME',axis =1)
TDCJ_Data.head()
```
